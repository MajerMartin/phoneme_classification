```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from subprocess import check_output
from IPython.display import display

%matplotlib inline
```

```python
def filter_logs(filters, logs):
    if not filters:
        return logs
    
    filtered_logs = []
    
    for flt in filters:
        for log in logs:
            if flt in log:
                filtered_logs.append(log)
          
    return filtered_logs
```

```python
DATASET = "SkodaAuto_25_10_mfcc"
FILTERS = []

logs = glob.glob(os.path.join("logs", "csv", DATASET, "*.csv"))
results_path = os.path.join("results", DATASET)
weights_path = os.path.join("weights", DATASET)

logs = filter_logs(FILTERS, logs)
```

```python
val_loss_df = pd.DataFrame()

for log in logs:
    model = os.path.basename(log)[:-4]
    print(log)
    df = pd.read_csv(log)
    val_loss_df[model] = df.val_loss
    
    try:
        with open(os.path.join(weights_path, model + "_metadata.txt")) as meta:
            ignore = True
            for line in meta.readlines():
                if line.startswith("overlap"):
                    print(line.strip())
                
                if line.startswith("Layer"):
                    ignore = False
                    
                if not ignore:
                    print(line.strip())
    except:
        print("NO METADATA")
    
    print()
    
    if "acc" in df.columns:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
        fig.suptitle(model)

        df.acc.plot(ax=ax1)
        df.val_acc.plot(ax=ax1)
        ax1.set_xlabel("epoch")
        ax1.set_ylabel("accuracy")
        ax1.legend(["train", "val"], loc="lower right")
    else:
        fig, ax2 = plt.subplots(figsize=(7, 4))
        ax2.set_title(model)
    
    try:
        df.loss.plot()
        df.val_loss.plot()
        ax2.set_xlabel("epoch")
        ax2.set_ylabel("loss")
        ax2.legend(["train", "val"], loc="upper right")

        plt.show()
    except:
        pass
    
    print()
    
    try:
        cmd = "HResults -I {} {} {}".format(os.path.join(results_path, model, "ref.mlf"),
                                            os.path.join(results_path, model, "phonemes"),
                                            os.path.join(results_path, model, "pred.mlf"))

        print(check_output(cmd.split(" ")).decode())
    except:
        print("NO RESULTS")
        
    print("\n")
    print("_" * 110)
    print("\n" * 3)
```

```python
val_loss_df[[col for col in val_loss_df.columns if "CTC" not in col]].plot(figsize=(10, 8))
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("validation loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.set_yscale("log")
val_loss_df[[col for col in val_loss_df.columns if "CTC" in col]].plot(ax=ax)
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("validation loss")
plt.xlabel("epoch")
plt.ylabel("log(loss)")
plt.show()
```
